```python
import mne
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
from PIL import Image
import io



# Function to read EDF file and extract signals using MNE
def read_edf_mne(file_path):
    raw = mne.io.read_raw_edf(file_path, preload=True, stim_channel=None)
    raw.notch_filter(freqs=[50, 100], picks='all')  # Apply notch filter to remove mains electricity noise
    data, times = raw[:]
    return data, raw.ch_names, times

# List of files with their labels
files_labels = [
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h01.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h02.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h03.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h04.edf', 'healthy'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h05.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h06.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h07.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h08.edf', 'healthy'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h09.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h10.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h11.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h12.edf', 'healthy'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h13.edf', 'healthy'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/h14.edf', 'healthy'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s01.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s02.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s03.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s04.edf', 'patient'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s05.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s06.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s07.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s08.edf', 'patient'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s09.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s10.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s11.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s12.edf', 'patient'),
    ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s13.edf', 'patient'), ('/content/drive/MyDrive/Schizophrenic/Dataset_Schizophrenia/s14.edf', 'patient')
]

# Function to read and extract signals from multiple EDF files
def extract_signals_from_files(files_labels):
    all_signals = []
    all_labels = []
    all_times = []
    all_signal_labels = []
    for file_path, label in files_labels:
        signals, signal_labels, times = read_edf_mne(file_path)
        all_signals.append(signals)
        all_labels.append(label)
        all_times.append(times)
        all_signal_labels.append(signal_labels)
    return all_signals, all_labels, all_times, all_signal_labels

# Extract signals from all files
all_signals, all_labels, all_times, all_signal_labels = extract_signals_from_files(files_labels)

# Function to convert signals to spectrograms and store them in memory
def generate_spectrogram_images(signals, labels, times, signal_labels, fs=256):
    spectrogram_images = []
    spectrogram_labels = []
    for i in range(len(signals)):
        for j in range(signals[i].shape[0]):
            f, t, Sxx = spectrogram(signals[i][j], fs)
            plt.figure(figsize=(10, 4))
            plt.pcolormesh(t, f, 10 * np.log10(Sxx))
            plt.ylabel('Frequency [Hz]')
            plt.xlabel('Time [sec]')
            plt.title(f'Spectrogram - {signal_labels[i][j]} ({labels[i]})')
            plt.colorbar(label='dB')

            # Save the image to an in-memory file
            buf = io.BytesIO()
            plt.savefig(buf, format='png')
            buf.seek(0)
            img = Image.open(buf).convert('RGB')  # Convert to RGB
            spectrogram_images.append(img)
            spectrogram_labels.append(labels[i])
            plt.close()
    return spectrogram_images, spectrogram_labels

# Generate spectrogram images and labels
spectrogram_images, spectrogram_labels = generate_spectrogram_images(all_signals, all_labels, all_times, all_signal_labels)

# Display some of the generated spectrogram images
selected_image = spectrogram_images[:2] + spectrogram_images[270:273]
selected_lables = spectrogram_labels[:2] + spectrogram_labels[270:273]

# print(f'images are {len(spectrogram_images)}, and labels are {len(spectrogram_labels)}')

for img, label in zip(selected_image, selected_lables):
    plt.figure()
    plt.imshow(img)
    plt.title(label)
    plt.axis('off')
    plt.show()
```

```python
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import EfficientNetB0
import tensorflow as tf
import numpy as np

# Convert PIL images to numpy arrays
spectrogram_images_np = np.array([np.array(img) for img in spectrogram_images])
spectrogram_labels_np = np.array([0 if label == 'healthy' else 1 for label in spectrogram_labels])

# Split dataset into training and testing sets
healthy_indices = np.where(spectrogram_labels_np == 0)[0]
patient_indices = np.where(spectrogram_labels_np == 1)[0]

test_indices = np.concatenate([healthy_indices[:53], patient_indices[:53]])
train_indices = np.concatenate([healthy_indices[53:], patient_indices[53:]])

train_images, test_images = spectrogram_images_np[train_indices], spectrogram_images_np[test_indices]
train_labels, test_labels = spectrogram_labels_np[train_indices], spectrogram_labels_np[test_indices]

# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels))

# Define image transformations and batch size
def preprocess(image, label):
    image = tf.image.resize(image, (224, 224))
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

batch_size = 32
train_dataset = train_dataset.map(preprocess).batch(batch_size).shuffle(buffer_size=len(train_dataset))
test_dataset = test_dataset.map(preprocess).batch(batch_size)


# Load the pre-trained EfficientNetB0 model
base_model = EfficientNetB0(include_top=False, input_shape=(224, 224, 3), weights='imagenet')
base_model.trainable = True

# Define the model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation='relu'),
    #layers.Dropout(0.2),
    layers.Dense(2, activation='softmax')
])

# Compile the model
model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model and save the history
history = model.fit(train_dataset, epochs=10, validation_data=test_dataset)
model.save('/content/transformer_model.h5')
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(test_dataset)
print(f'Test accuracy: {accuracy * 100:.2f}%')

predicts = model.predict(train_dataset)
print(predicts)
```

```python

import matplotlib.pyplot as plt
# Plot accuracy and loss
def plot_training_history(history):
    # Plot accuracy
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Call the function to plot the training history
plot_training_history(history)
```
